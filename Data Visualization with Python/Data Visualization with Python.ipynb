# **MatPlotLib Plot**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_json(r'/rain.json')
```

```python
plt.figure(figsize=(15,5))
plt.plot(df['Month'], df['Temperature'], label='Temperature')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df['Month'], df['Rainfall'], label='Rainfall')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df['Month'], df['Temperature'], label='Temperature')
plt.plot(df['Month'], df['Rainfall'], label='Rainfall')
plt.legend()
plt.show()
```

# **Scatterplot Graph**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'/tempYearly.csv')
print(df)
```

```python
sns.set(rc={'figure.figsize':(12,6)})
sns.scatterplot(x='Year', y='Temperature', data=df)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(12,6)})
sns.regplot(x='Year', y='Temperature', data=df)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(12,6)})
sns.regplot(x='Rainfall', y='Temperature', data=df)
plt.show()
```

# **Heatmap plot**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'/birthYearly.csv')
print(data)
```

```python
dataP = data.pivot(index="month", columns="year",values="births")
print(dataP)
```

```python
sns.heatmap(dataP, annot=True, fmt="d")
plt.show()
```

# **Jointplot Distribution**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv(r'/tempYearly.csv')
print(df)
```

```python
sns.jointplot(x="Rainfall",y="Temperature", data=df)
plt.show( )
```

```python
sns.jointplot(x="Rainfall",y="Temperature", data=df, kind='hex')
plt.show( )
```

```python
sns.jointplot(x="Rainfall",y="Temperature", data=df, kind='reg')
plt.show( )
```
